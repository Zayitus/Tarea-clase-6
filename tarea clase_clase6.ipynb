### Ejercicio 1

```python
def calcular_imc(peso_kg, altura_cm):
    # Convertir la altura de cm a metros
    altura_m = altura_cm / 100
    # Calcular el IMC
    imc = peso_kg / (altura_m ** 2)
    return imc


def main():
    # Solicitar peso y altura al usuario
    peso = float(input("Introduce tu peso en kg: "))
    altura = float(input("Introduce tu altura en cm: "))

    # Calcular el IMC
    imc = calcular_imc(peso, altura)

    # Mostrar el resultado
    print(f"Tu IMC es: {imc:.2f}")


if __name__ == "__main__":
    main()
```

### Ejercicio N 2 

```python
def analizar_numero(numero):
    # Crear un diccionario para almacenar los resultados
    resultado = {
        "paridad": "",
        "signo": "",
        "divisible_por_3": False,
        "divisible_por_5": False,
        "divisible_por_3_y_5": False
    }

    # Determinar si el número es par o impar
    if numero % 2 == 0:
        resultado["paridad"] = "par"
    else:
        resultado["paridad"] = "impar"

    # Determinar si el número es positivo, negativo o cero
    if numero > 0:
        resultado["signo"] = "positivo"
    elif numero < 0:
        resultado["signo"] = "negativo"
    else:
        resultado["signo"] = "cero"

    # Determinar si el número es divisible por 3
    if numero % 3 == 0:
        resultado["divisible_por_3"] = True

    # Determinar si el número es divisible por 5
    if numero % 5 == 0:
        resultado["divisible_por_5"] = True

    # Determinar si el número es divisible por 3 y por 5
    if numero % 3 == 0 and numero % 5 == 0:
        resultado["divisible_por_3_y_5"] = True

    return resultado


def mostrar_caracteristicas(numero, informacion):
    caracteristicas = []

    caracteristicas.append(f"El número {numero} es {informacion['paridad']}")
    caracteristicas.append(f"El número {numero} es {informacion['signo']}")
    if informacion["divisible_por_3"]:
        caracteristicas.append(f"El número {numero} es divisible por 3")
    if informacion["divisible_por_5"]:
        caracteristicas.append(f"El número {numero} es divisible por 5")
    if informacion["divisible_por_3_y_5"]:
        caracteristicas.append(
            f"El número {numero} es divisible por 3 y por 5")

    return caracteristicas


# Ejemplo de uso
numero = int(input("Introduce un número entero: "))
informacion = analizar_numero(numero)
caracteristicas = mostrar_caracteristicas(numero, informacion)

print("Características del número analizado:")
for caracteristica in caracteristicas:
    print(caracteristica)
```

#### Opcional ejercicio 1 con funcion lambda

```python
def analizar_numero(numero):
    # Crear un diccionario para almacenar los resultados
    resultado = {
        "paridad": (lambda x: "par" if x % 2 == 0 else "impar")(numero),
        "signo": (lambda x: "positivo" if x > 0 else ("negativo" if x < 0 else "cero"))(numero),
        "divisible_por_3": (lambda x: x % 3 == 0)(numero),
        "divisible_por_5": (lambda x: x % 5 == 0)(numero),
        "divisible_por_3_y_5": (lambda x: x % 3 == 0 and x % 5 == 0)(numero)
    }

    return resultado


def mostrar_caracteristicas(numero, informacion):
    caracteristicas = []

    caracteristicas.append(f"El número {numero} es {informacion['paridad']}")
    caracteristicas.append(f"El número {numero} es {informacion['signo']}")
    if informacion["divisible_por_3"]:
        caracteristicas.append(f"El número {numero} es divisible por 3")
    if informacion["divisible_por_5"]:
        caracteristicas.append(f"El número {numero} es divisible por 5")
    if informacion["divisible_por_3_y_5"]:
        caracteristicas.append(
            f"El número {numero} es divisible por 3 y por 5")

    return caracteristicas


# Ejemplo de uso
numero = int(input("Introduce un número entero: "))
informacion = analizar_numero(numero)
caracteristicas = mostrar_caracteristicas(numero, informacion)

print("Características del número analizado:")
for caracteristica in caracteristicas:
    print(caracteristica)
```

### Ejercicio 3

```python
import math


def area_cuadrado(lado):
    return lado ** 2


def perimetro_cuadrado(lado):
    return 4 * lado


def area_triangulo_equilatero(lado):
    return (math.sqrt(3) / 4) * lado ** 2


def perimetro_triangulo_equilatero(lado):
    return 3 * lado


def area_triangulo_rectangulo(base, altura):
    return (base * altura) / 2


def perimetro_triangulo_rectangulo(base, altura):
    hipotenusa = math.sqrt(base ** 2 + altura ** 2)
    return base + altura + hipotenusa


def area_triangulo_isosceles(base, lado):
    altura = math.sqrt(lado ** 2 - (base / 2) ** 2)
    return (base * altura) / 2


def perimetro_triangulo_isosceles(base, lado):
    return 2 * lado + base


def area_circulo(radio):
    return math.pi * radio ** 2


def perimetro_circulo(radio):
    return 2 * math.pi * radio


def area_rectangulo(base, altura):
    return base * altura


def perimetro_rectangulo(base, altura):
    return 2 * (base + altura)


def main():
    while True:
        print("\nSeleccione una figura geométrica:")
        print("1. Cuadrado")
        print("2. Triángulo Equilátero")
        print("3. Triángulo Rectángulo")
        print("4. Triángulo Isósceles")
        print("5. Círculo")
        print("6. Rectángulo")
        print("7. Salir")
        opcion_figura = input("Ingrese el número de la figura geométrica: ")

        if opcion_figura == '7':
            print("Saliendo del programa...")
            break

        print("\n¿Qué desea calcular?")
        print("1. Área")
        print("2. Perímetro")
        opcion_calculo = input("Ingrese el número de la opción: ")

        if opcion_figura == '1':  # Cuadrado
            lado = float(input("Ingrese el lado del cuadrado: "))
            if opcion_calculo == '1':
                print(f"Área del cuadrado: {area_cuadrado(lado)}")
            elif opcion_calculo == '2':
                print(f"Perímetro del cuadrado: {perimetro_cuadrado(lado)}")

        elif opcion_figura == '2':  # Triángulo Equilátero
            lado = float(input("Ingrese el lado del triángulo equilátero: "))
            if opcion_calculo == '1':
                print(
                    f"Área del triángulo equilátero: {area_triangulo_equilatero(lado)}")
            elif opcion_calculo == '2':
                print(
                    f"Perímetro del triángulo equilátero: {perimetro_triangulo_equilatero(lado)}")

        elif opcion_figura == '3':  # Triángulo Rectángulo
            base = float(input("Ingrese la base del triángulo rectángulo: "))
            altura = float(
                input("Ingrese la altura del triángulo rectángulo: "))
            if opcion_calculo == '1':
                print(
                    f"Área del triángulo rectángulo: {area_triangulo_rectangulo(base, altura)}")
            elif opcion_calculo == '2':
                print(
                    f"Perímetro del triángulo rectángulo: {perimetro_triangulo_rectangulo(base, altura)}")

        elif opcion_figura == '4':  # Triángulo Isósceles
            base = float(input("Ingrese la base del triángulo isósceles: "))
            lado = float(input("Ingrese el lado del triángulo isósceles: "))
            if opcion_calculo == '1':
                print(
                    f"Área del triángulo isósceles: {area_triangulo_isosceles(base, lado)}")
            elif opcion_calculo == '2':
                print(
                    f"Perímetro del triángulo isósceles: {perimetro_triangulo_isosceles(base, lado)}")

        elif opcion_figura == '5':  # Círculo
            radio = float(input("Ingrese el radio del círculo: "))
            if opcion_calculo == '1':
                print(f"Área del círculo: {area_circulo(radio)}")
            elif opcion_calculo == '2':
                print(f"Perímetro del círculo: {perimetro_circulo(radio)}")

        elif opcion_figura == '6':  # Rectángulo
            base = float(input("Ingrese la base del rectángulo: "))
            altura = float(input("Ingrese la altura del rectángulo: "))
            if opcion_calculo == '1':
                print(f"Área del rectángulo: {area_rectangulo(base, altura)}")
            elif opcion_calculo == '2':
                print(
                    f"Perímetro del rectángulo: {perimetro_rectangulo(base, altura)}")

        else:
            print("Opción no válida. Por favor, intente de nuevo.")


if __name__ == "__main__":
    main()
```
